## AdaLine:

```python

x1 = np.random.normal(0, 0.1, 100)
y1 = np.random.normal(0, 0.4, 100)
x2 = np.random.normal(1, 0.2, 100)
y2 = np.random.normal(1, 0.2, 100)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f'AdaLine')
p1=plt.scatter(x1, y1, color="red")
p2=plt.scatter(x2, y2, color="blue")
plt.legend((p1, p2),('Class 1', 'Class 2'))
plt.show()
```

```python
def adalineModel(X1,Y1,X2,Y2):
  er=[]
  w_x=0
  w_y=0
  b=1
  a=0.01
  ln=len(X1)+len(X2)
  for j in range(500):  #epoch=500
     M=0
     for i in range(ln):
         t=0
         xi1=0
         xi2=0
         if i%2==0:
           xi12=X1[int(i/2)]
           xi22=Y1[int(i/2)]
         else:
          xi11=X2[int(math.floor(i/2))]
          xi21=Y2[int(math.floor(i/2))]
         if i%2!=0:
              trgt=-1
              net =w_x*xi11+ w_y*xi21+b
              w_x =w_x+a*(trgt-net)*xi11
              w_y =w_y+a*(trgt-net)*xi21
              b =b+a*(trgt-net)
         else:
             trgt=1
             net =w_x*xi12+w_y*xi22+b
             w_x =w_x+a*(trgt-net)*xi12
             w_y =w_y+a*(trgt-net)*xi22
             b =b+a*(trgt-net)

         error=(1/2)*(trgt-net)**2
         if M<error:
             M=error
     er.append(M)
  return er ,w_x ,b,w_y
```

```python
err,wx,bs,wy = adalineModel(x1,y1,x2,y2)
plt.ylabel("error")
plt.xlabel("epochs")
plt.title('Error')
plt.plot(err, color="blue")
plt.grid()
plt.show()
```

```python
p1=plt.scatter(x1, y1, c="red")
p2=plt.scatter(x2, y2, c="blue")
xd = np.array([-0.3, 1.5])
L = (wx * xd + bs) / (-wy)
plt.plot(xd, L, "k", lw=1, ls="--",alpha=0.7)
plt.title(f'AdaLine')
plt.xlabel("x")
plt.ylabel("y")
plt.legend((p1, p2),('Class 1', 'Class 2'))
plt.show()
```

```python
x21 = np.random.normal(0, 0.4, 100)
y21 = np.random.normal(0, 0.4, 100)

x22 = np.random.normal(1, 0.3, 100)
y22 = np.random.normal(1, 0.3, 100)

pp1=plt.scatter(x21, y21, c="red")
pp2=plt.scatter(x22, y22, c="blue")
plt.title(f'AdaLine')
plt.xlabel("x")
plt.ylabel("y")
plt.legend((pp1, pp2),('Class 1', 'Class 2'))
plt.show()
```
